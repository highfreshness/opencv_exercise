```python
import cv2
```

```python
image = cv2.imread("test.jpg")
```

```python
print("width : {} pixels".format(image.shape[1]))
```

```python
print("height : {} pixels".format(image.shape[0]))
```

```python
print("channels : {} ".format(image.shape[2]))
```

```python
(b,g,r) = image[10,10]
print("Pixel at(10,10) - R{}, G{}, B{}".format(r, g, b))
```

```python
image[10,10] = (0,0,255)
(b,g,r) = image[10,10]
print("Pixel at(10,10) - R{}, G{}, B{}".format(r, g, b))
```

```python
corner = image[0:100, 0:200]
cv2_imshow(corner)
image[0:100, 0:200] = (0,255,0)
```

```python
# cv2.imshow("image", image)
cv2_imshow(image)
```

```python
# cv2.waitKey(0)
```

```python
import numpy as np
green = (0,255,0)
canvas = np.zeros_like(image, )
```

```python
cv2.line(canvas, (0,0), (300, 200), green)
cv2_imshow(canvas)
```

```python
cv2.rectangle(canvas, (10, 20), (60, 80), green)
cv2_imshow(canvas)
```

```python
red = (0,0,255)
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
cv2_imshow(canvas)
```

```python
blue = (255,0,0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
cv2_imshow(canvas)
```
